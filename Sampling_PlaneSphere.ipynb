# Sampling two random vectors

```python
v1=(SPI.S[1][0],SPI.S[1][1],SPI.S[1][2])
v2=(SPI.S[2][0],SPI.S[2][1],SPI.S[2][2])
```

```python
print(v1[0])
```

# Matrix for Scaling

```python
parameter1 = np.transpose(np.random.randn(1000)[np.newaxis])
parameter2 = np.transpose(np.random.randn(1000)[np.newaxis])
```

# Scaling the two vectors to obtain points on a plane

```python
sphere_plane=parameter1*v1+parameter2*v2 
```

```python
sphere_plane[1]
```

# Visual

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from numpy.random import rand
from pylab import figure


SS=SPI.S# m is an array of (x,y,z) coordinate triplets
LS=SPS.sphere_line_OS
SP=sphere_plane
#print(LS)
#n=exits_sorted['name']
#print(m)

fig=figure(figsize=(30,15))
##fig = figure()
ax = Axes3D(fig)


for i in range(len(LS)): #plot each point + it's index as text above
    ax.scatter(LS[i][0],LS[i][1],LS[i][2],color='b')
    
for i in range(len(SS)): #plot each point + it's index as text above
    ax.scatter(SS[i][0],SS[i][1],SS[i][2],color='r')
    
for i in range(len(SP)): #plot each point + it's index as text above
    ax.scatter(SP[i][0],SP[i][1],SP[i][2],color='g')



ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
pyplot.show()
```
